```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

### COUNTPLOT ###

```python
sns.countplot(x= df['sex'])
```

```python
sns.countplot(x=df['sex'], hue= df['smoker'])
```

### BARPLOT ### 

```python
sns.barplot(x = df['sex'],y = df['tip'])
```

```python
sns.barplot(x = df['sex'],y = df['tip'],estimator=np.sum)
```

### BOXPLOT ###

```python
sns.boxplot(x='tip',y='day',data =df,palette='rainbow')
```

### VIOLENPLOT ###

```python
sns.violinplot(x='day',y='tip',data=df,palette='rainbow')
sns.swarmplot(x='day',y='tip',data = df)
```

#### STRIPPLOT AND SWARMPLOT ###

```python
sns.violinplot(x='day',y='tip',data=df,palette='rainbow')
sns.stripplot(x = 'tip',y ='day',data = df)
```

```python
sns.violinplot(x='day',y='tip',data=df,palette='rainbow')
sns.stripplot(y = 'tip',x ='day',data = df)
```

```python
sns.violinplot(x='day',y='tip',data=df,palette='rainbow')
sns.swarmplot(x='day',y='tip',data = df)
```

```python
sns.swarmplot(x='tip',y='day',data = df)
```
