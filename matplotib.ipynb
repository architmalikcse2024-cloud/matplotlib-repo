```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

### MATPPLOTLIB BASICS ###

```python
x = np.linspace(0,5,11)
x
y= x**2
y
```

```python
plt.plot(x,y)
plt.title("hello")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

#### MATPLOTLIB SUBPLOTS ###

```python
plt.subplot(2,2,1)
plt.plot(x,y**3)
plt.subplot(2,2,2)
plt.plot(x,x)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
```

### OBJECT ORIENTED WAY ###

```python
fig = plt.figure(figsize=(8,8),dpi=(100))
axis1 = fig.add_axes([0.1,0.1,1.0,1.0])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.5,.5,.5])
axis2.plot(x,y)
```

### TYPES OF PLOTS ###

### SAVING A PLOT ###

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basic plot.png")
```
