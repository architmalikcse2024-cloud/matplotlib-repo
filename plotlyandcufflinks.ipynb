```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cufflinks as cf
import plotly.graph_objects as go
```

```python
cf.go_offline()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
import plotly.graph_objects as go

fig = go.Figure(
    data=[go.Scatter(y=tips["total_bill"], mode="lines+markers", name="total_bill")]
)
fig.update_layout(
    title="Total Bill",
    xaxis_title="Index",
    yaxis_title="Total Bill"
)
fig.show()
```

```python
avg_tip_by_day = tips.groupby('day')['tip'].mean()

fig = go.Figure(
    data=[go.Bar(
        x=avg_tip_by_day.index,
        y=avg_tip_by_day.values,
        marker_color='orange'
    )]
)

fig.update_layout(
    title='Average Tip by Day',
    xaxis_title='Day',
    yaxis_title='Average Tip'
)

fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=tips["total_bill"],
    y=tips["tip"],
    mode="markers",
    marker=dict(color="orange")
))

fig.update_layout(
    title="Total Bill vs Tip",
    xaxis_title="Total Bill",
    yaxis_title="Tip"
)

fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Box(
    x=tips["total_bill"],
    y=tips["day"],
    name="Total Bill by Day"
))
fig.show()
```
